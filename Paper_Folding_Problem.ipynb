# Problem 1: Implementation using exponentiation arithmetic operator

```python
# Code to calculate the thickness when the paper is folded forty-three times
THICKNESS = 0.00008
folded_thickness = THICKNESS * 2 ** 43

print("Thickness: {:,.2f} meters".format(folded_thickness))
```

# Problem 2: Unit Conversion

```python
# Convert meters to kilometers and display with two decimal places
thickness_kilometer = folded_thickness / 1000
print("Thickness: {:,.2f} kilometers".format(thickness_kilometer))

thickness_relative_10000k = thickness_kilometer/10000
print("Thickness: {:,.2f}".format(thickness_relative_10000k))
```

# Problem 3 : Create using a for statement

```python
folded_thickness_2 = THICKNESS
folded_times = 43
for i in range(1, folded_times + 1):  #We use a for to repeat the
  folded_thickness_2 *= 2             #operation as many times as required

print("Thickness: {:,.2f} meters".format(folded_thickness_2))
```

# Problem 4: Comparison of calculation time

```python
import time
start = time.time()

# We copy the code that uses "for"
folded_thickness_2 = THICKNESS
folded_times = 43
for i in range(1, folded_times + 1):
  folded_thickness_2 *= 2

print("Thickness: {:,.2f} meters".format(folded_thickness_2))

elapsed_time = time.time() - start    #We get the time it takes to execute the code
print("time : {}[s]".format(elapsed_time))
```

# Problem 5: Saving to a list

```python
THICKNESS = 0.00008
num_of_folds = 43

folded_thickness_values = []    #Create a list for the values

folded_thickness_5 = THICKNESS
folded_thickness_values.append(folded_thickness_5) #Add the first value to the list

for i in range(43):
  folded_thickness_5 *= 2
  folded_thickness_values.append(folded_thickness_5)  #We add the values while we are operating with the for

print(len(folded_thickness_values)) #Print with “len” to find out how many values are in the list
```

# Problem 6: Displaying a line graph

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Display the graph. Title and axis label name.
plt.title("Thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_values)  # We enter “folded_thickness_values” in the plot function to plot the list we have already made
plt.show()
```

# Problem 7: Customizing graphs

```python
# Display a purple line graph and fontsize for the title and axis label name
plt.title("Thickness of folded paper", fontsize=20)
plt.xlabel("number of folds", fontsize=14)
plt.ylabel("thickness [m]", fontsize=14)
plt.tick_params(labelsize=14)

# We enter “folded_thickness_values” in the plot function to plot the list we have already made
# In addition, we add markers, line sizes and legends
plt.plot(folded_thickness_values, color='purple', linewidth=2, linestyle="--", marker="*", markersize="7", label="Problem 7: Customizing graph")

plt.legend() #Activate the changes in the graphs
plt.show()
```
